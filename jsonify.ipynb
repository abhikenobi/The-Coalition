# Concatenation

* Combining all Years into single DataFrame

```python
frames = [y2001_df, y2002_df, y2003_df, y2004_df, y2005_df, y2006_df, y2007_df, y2008_df, y2009_df,
          y2010_df, y2011_df, y2012_df, y2013_df, y2014_df, y2015_df, y2016_df, y2017_df, y2018_df]
combined_df = pd.concat(frames, ignore_index=True)
combined_df
```

```python
columnn_names = ['Plant_ID','Nameplate','Technology','Latitude','Longitude','State','Year']
combined_df = combined_df.reindex(columns=columnn_names)
combined_df
```

# Exporting to SQL Database

* Using sqlalchemy and Postgres to export all DataFrames into their own tables in a database

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/energy_db')

engine.table_names()
```

```python
solar_df.to_sql(
    name = "solar",
    schema = 'public',
    con = engine,
    if_exists = 'replace',
    index = True
    )
```
